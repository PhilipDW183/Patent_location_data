```python
import pandas as pd
import numpy as np
```

This workbook explores the dataset from here: https://www.nature.com/articles/s41597-019-0264-6

This combines the first and subsequent variables into one and then pushes the results

```python
#read in teh first first and subsequent dataset
first_subsequent1 = pd.read_csv("Patents data/first_and_subsequent_filings_1.txt", sep = ",", header = 0)
```

```python
#read in the second first and subsequent dataset
first_subsequent2 = pd.read_csv("Patents data/first_and_subsequent_filings_2.txt", sep = ",", header = 0)
```

```python
#vertically merge the two
first_and_subsequent = pd.concat([first_subsequent1, first_subsequent2])
```

```python
#check the results
first_and_subsequent
```

```python
#extract results that come from the EP and WO
first_and_subsequent_EP = first_and_subsequent[first_and_subsequent["publn_auth"] == "EP"]
first_and_subsequent_WO = first_and_subsequent[first_and_subsequent["publn_auth"] == "WO"]
```

```python
#output the results to seperate csvs
first_and_subsequent_EP.to_csv("Patents data/first_and_subsequent_EP.csv")
first_and_subsequent_WO.to_csv("Patents data/first_and_subsequent_WO.csv")
```
